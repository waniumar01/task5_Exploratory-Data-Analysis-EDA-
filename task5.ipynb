```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train.csv")
df.head()
```

```python
df['Survived'].value_counts()
```

```python
df['Sex'].value_counts()
```

```python
df['Age'].hist()
plt.title("Age Distribution")
plt.show()
```

**Observation:**
Most passengers were between 15 to 40 years old.


```python
sns.boxplot(x=df['Fare'])
plt.title("Fare Distribution")
plt.show()
```

**Observation:**
Most passengers paid low fares, while a few passengers paid very high fares, resulting in a right-skewed distribution with outliers.


```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

**Observation:**
Female passengers had a significantly higher survival rate than males.


```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare")
plt.show()
```

**Observation:**
Passengers who paid higher fares had a higher chance of survival.


```python
sns.pairplot(df[['Age','Fare','Survived']])
plt.show()
```

**Observations:**

1. Passengers who paid higher fares had a higher probability of survival.
2. Age does not show a strong linear relationship with survival.
3. Fare values are highly skewed, with most passengers paying lower fares.
4. Multivariate analysis indicates fare is a more influential factor for survival compared to age.




```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(8,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**Observation:**
Fare shows a positive correlation with survival, while Pclass has a negative correlation, indicating higher survival chances for higher-class passengers.


## Summary of Findings

- Most passengers were aged between 15 to 40 years.
- Fare distribution is highly right-skewed with several high-value outliers.
- Female passengers had a significantly higher survival rate than males.
- Passengers who paid higher fares had better chances of survival.
- Passenger class shows a negative correlation with survival, indicating higher survival rates for first-class passengers.
- Age shows only a weak correlation with survival.
- Multivariate analysis highlights fare and passenger class as important factors influencing survival.
